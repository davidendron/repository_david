# Proteomics data Class

```python
class Proteomic_data:
    """
    
    
    """
    data_excel = "schmidt_data.xlsx"
    table_page = "Table S6"
    
    def __init__(self, remove_gene_na=True):
        """
        
        """
        import pandas as pd
        self.proteomic_data_table = pd.read_excel(self.data_excel, sheet_name=self.table_page ,skiprows=2)
        print(f"Loaded {self.table_page} from {self.data_excel}.")
        
        if remove_gene_na:
            self.proteomic_data_table_geneNaNFiltered = self.proteomic_data_table.dropna(subset=['Gene'])
            print('Genes NaN removed: {}.'.format(len(set(self.proteomic_data_table['Gene']).difference(self.proteomic_data_table_geneNaNFiltered['Gene']))))
                  
        self.table_columns = list(self.proteomic_data_table.columns)
        
    def genes_filter(self, string, column="Gene"):
        """
    
        """
        selection_mask = self.proteomic_data_table_geneNaNFiltered[column].str.contains(string)
        return self.proteomic_data_table_geneNaNFiltered[selection_mask]
    
    def uniprot_request(self):
        """
    
        """
        import requests, sys
        UNIPROT_API = "https://www.ebi.ac.uk/proteins/api"
        PROTEOMICS_ENDPOINT = "/proteins"
        request_params = {
            'offset':'0',
            'size':'100',
            'accession':'P09546'
            }

        request = requests.get(f"{UNIPROT_API}{PROTEOMICS_ENDPOINT}", params=request_params, headers={ "Accept" : "application/json"})

        if not request.ok:
            request.raise_for_status()
            sys.exit()

        responseBody = request.text
        
        response = eval(responseBody.replace('false', 'False'))[0]
        return response.keys()
        
    def save(self):
        # TODO: Terminar una función para guardar las tablas generadas.
        return
```
